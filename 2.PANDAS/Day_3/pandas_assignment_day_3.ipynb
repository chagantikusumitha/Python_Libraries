###### 17. Write a Pandas program to get the current date, oldest date and number of days between Current date and oldest date of Ufo dataset.

```python
df[['date','time']] = df['Date_time'].str.split(expand = True)
df['Date_time'] = (pd.to_datetime(df['date']) + pd.to_timedelta(df.pop('time')+':00'))
df['Date_time'] = df['Date_time'].astype('datetime64[ns]')
```

```python
#### code here
print("current date",df["Date_time"].max())
print("oldest date",df["Date_time"].min())
print((df["Date_time"].max() - df["Date_time"].min()).days)
```
